## $\color{blue}{\text{Hurts exponent}}$

In this plot we are going to look up the evolution of the Hurst exponent, wich is a measure of the long-term memory of time series and permits evaluate the physical phenomena according to:

* $H\leq0.5$: In this case we find an **antipersistent** time series this means that the temporal serie varies gradually and the best la prediction is the mean of the last elements.

* $H=0.5$: The serie has an **estochatics** behaviour and it is extremelly difficult predict its future bahaviour.

* $H\geq0.5$:  The time series is **persistent**, reazon for which long-term period will be remain.
